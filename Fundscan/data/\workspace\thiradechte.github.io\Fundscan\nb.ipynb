```python
import pandas as pd
```

```python
cols_name = []
with open('cols_name.txt', 'r') as file:
    cols = file.read()
```

```python
cols_name = cols.split(",")
for i in range(0, len(cols_name)):
    cols_name[i] = cols_name[i][1:len(cols_name[i])-1]
    
```

```python
df = pd.read_csv('Fund Comparison.csv', names=cols_name)
```

```python
for i in range(len(df.columns)):
    df.iloc[:, i].replace(["-", "-%"], None, inplace=True)
    df.iloc[:, i].replace("ไม่มี", "No", inplace=True)
```

```python
df.to_csv('cleaned_data.csv')
```
